```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set_theme(color_codes=True)

df = pd.read_csv('Bank Customer Churn Prediction.csv')
print(df.head())
```

```python
#Counting 1 and 0 Value in Churn column
color_wheel = {1: "#0392cf", 2: "#7bc043"}
colors = df["churn"].map(lambda x: color_wheel.get(x + 1))
print(df.churn.value_counts())
p=df.churn.value_counts().plot(kind="bar")
```

```python
#Change value in country column
df['country'] = df['country'].replace(['Germany'],'0')
df['country'] = df['country'].replace(['France'],'1')
df['country'] = df['country'].replace(['Spain'],'2')
#Change value in gender column
df['gender'] = df['gender'].replace(['Female'],'0')
df['gender'] = df['gender'].replace(['Male'],'1')
df.head()
```

```python
#convert object data types column to integer
df['country'] = pd.to_numeric(df['country'])
df['gender'] = pd.to_numeric(df['gender'])
df.dtypes
```

```python
#Remove customer_id column
df2 = df.drop('customer_id', axis=1)
df2.head()
```

```python
sns.heatmap(df2.corr(), fmt='.2g')
```

**Build Machine Learning Model**


```python
X = df2.drop('churn', axis=1)
y = df2['churn']
#test size 20% and train size 80%
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,random_state=7)
```

**Decision Tree**

```python

from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
y_pred = dtree.predict(X_test)
print("Accuracy Score :", accuracy_score(y_test, y_pred)*100, "%")
```

Random Forest


```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
print("Accuracy Score :", accuracy_score(y_test, y_pred)*100, "%")
```

Support Vector Machine

```python
from sklearn import svm
svm = svm.SVC()
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
print("Accuracy Score :", accuracy_score(y_test, y_pred)*100, "%")
```

XGBoost

```python
from xgboost import XGBClassifier
xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train)
y_pred = xgb_model.predict(X_test)
print("Accuracy Score :", accuracy_score(y_test, y_pred)*100, "%")
```

```python
# prompt: create a table with different scores - Accuracy, Precision, Recall and F1 Score for the above 4 algorithms used

import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Create a dictionary to store the scores
scores = {
    'Algorithm': ['Decision Tree', 'Random Forest', 'SVM', 'XGBoost'],
    'Accuracy': [],
    'Precision': [],
    'Recall': [],
    'F1 Score': []
}

# Calculate the scores for each algorithm
scores['Accuracy'].append(accuracy_score(y_test, dtree.predict(X_test)))
scores['Precision'].append(precision_score(y_test, dtree.predict(X_test)))
scores['Recall'].append(recall_score(y_test, dtree.predict(X_test)))
scores['F1 Score'].append(f1_score(y_test, dtree.predict(X_test)))

scores['Accuracy'].append(accuracy_score(y_test, rfc.predict(X_test)))
scores['Precision'].append(precision_score(y_test, rfc.predict(X_test)))
scores['Recall'].append(recall_score(y_test, rfc.predict(X_test)))
scores['F1 Score'].append(f1_score(y_test, rfc.predict(X_test)))

scores['Accuracy'].append(accuracy_score(y_test, svm.predict(X_test)))
scores['Precision'].append(precision_score(y_test, svm.predict(X_test)))
scores['Recall'].append(recall_score(y_test, svm.predict(X_test)))
scores['F1 Score'].append(f1_score(y_test, svm.predict(X_test)))

scores['Accuracy'].append(accuracy_score(y_test, xgb_model.predict(X_test)))
scores['Precision'].append(precision_score(y_test, xgb_model.predict(X_test)))
scores['Recall'].append(recall_score(y_test, xgb_model.predict(X_test)))
scores['F1 Score'].append(f1_score(y_test, xgb_model.predict(X_test)))

# Create a DataFrame from the dictionary
scores_df = pd.DataFrame(scores)

# Print the DataFrame
print(scores_df)
```

XGBoost and Random Forest - Deep Dive

```python
#importing classification report and confusion matrix from sklearn

from sklearn.metrics import classification_report, confusion_matrix
```

Random Forest

```python
y_pred = rfc.predict(X_test)
print("Classification report - n", classification_report(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,5))
sns.heatmap(data=cm,linewidths=.5, annot=True,square = True,  cmap = 'Blues')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
all_sample_title = 'Accuracy Score: {0}'.format(rfc.score(X_test, y_test))
plt.title(all_sample_title, size = 15)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
y_pred_proba = rfc.predict_proba(X_test)[:][:,1]
df_actual_predicted = pd.concat([pd.DataFrame(np.array(y_test), columns=['y_actual']), pd.DataFrame(y_pred_proba, columns=['y_pred_proba'])], axis=1)
df_actual_predicted.index = y_test.index
fpr, tpr, tr = roc_curve(df_actual_predicted['y_actual'], df_actual_predicted['y_pred_proba'])
auc = roc_auc_score(df_actual_predicted['y_actual'], df_actual_predicted['y_pred_proba'])
plt.plot(fpr, tpr, label='AUC = %0.4f' %auc)
plt.plot(fpr, fpr, linestyle = '--', color='k')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve', size = 15)
plt.legend()
```

```python
# prompt: Explain the ROC curve above

The ROC curve is a graphical representation of the performance of a binary classifier system. It plots the true positive rate (TPR) against the false positive rate (FPR) at various threshold settings.

The TPR is the proportion of actual positive samples that are correctly identified as positive, while the FPR is the proportion of actual negative samples that are incorrectly identified as positive.

The ROC curve is a useful tool for evaluating the performance of a classifier because it provides a visual representation of how well the classifier is able to distinguish between positive and negative samples.

A perfect classifier would have a TPR of 1 and an FPR of 0, which would correspond to a point in the upper left corner of the ROC curve.

The ROC curve for the XGBoost model in the example above shows that the model has a high TPR and a low FPR, which indicates that it is performing well at distinguishing between positive and negative samples.

The AUC (area under the curve) is a single number that summarizes the performance of a classifier. It is the area under the ROC curve, and it ranges from 0 to 1.

A perfect classifier would have an AUC of 1, while a random classifier would have an AUC of 0.5.

The AUC for the XGBoost model in the example above is 0.97, which indicates that it is performing very well.
```
